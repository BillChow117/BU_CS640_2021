```python
import pandas as pd
print(pd.__version__)
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
from sklearn.metrics import classification_report
```

# Loading Dataset File

```python
train=pd.read_csv('training.1600000.processed.noemoticon.csv',encoding='latin1',header=None)
test=pd.read_csv('testdata.manual.2009.06.14.csv',encoding='latin1',header=None)
```

```python
train=train[[0,5]]
train.head()
```

```python
train[0].value_counts()
```

```python
test=test[[0,5]]
test.head()
```

```python
test[0].value_counts()
```

```python
regexMap={r"<[\w'/'\s]*>": "",r"[\'\"\-]+": "",r"@[\w]+":"",r"https?:\/\/\S+":""}
def preprocess(datainput):
    t=datainput
    for regx in regexMap.keys():
        t = re.sub(regx, regexMap[regx], t)
    return t
```

```python
train[5]=train[5].apply(preprocess)
train[0].value_counts()
```

```python
test=test.drop(test[test[0]==2].index)
test[5]=test[5].apply(preprocess)
test[0].value_counts()
```

# Fit Count Vectorizer on training data

```python
vectorizer = CountVectorizer()
```

## Stopwords

```python
stop_words_vectorizer=CountVectorizer(stop_words='english')
stop_words_vectorizer.fit(train[5].values)
```

```python
stop_words_vectorizer.vocabulary_
```

```python
x_input=stop_words_vectorizer.transform(train[5].values)
x_test_input=stop_words_vectorizer.transform(test[5].values)
```

```python
nb = MultinomialNB()
```

```python
nb.fit(x_input,train[0])
```

```python
y_pred_class = nb.predict(x_test_input)
metrics.accuracy_score(test[0].values, y_pred_class)
```

```python
print(classification_report(y_true=test[0].values,y_pred=y_pred_class))
```
